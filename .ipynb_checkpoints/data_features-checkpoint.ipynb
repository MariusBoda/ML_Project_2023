```python
import numpy as np                   
import pandas as pd                  
import matplotlib.pyplot as plt


data = pd.read_csv("Apple-Twitter-Sentiment-DFE.csv", sep=';', encoding='utf-8', header = 0)

headers = data.iloc[0]
data.columns = headers
data = data.iloc[1:]

# Specify the column labels you want to remove
columns_to_remove = data.columns[0:5].tolist() + data.columns[7:11].tolist()

# Remove the specified columns
data = data.drop(columns=columns_to_remove)

filtered_data = data[
    (data['sentiment'].isin([1, 3, 5])) |  # Keep rows with sentiment values 1, 3, or 5
    (data['text'].str.strip() != '')  # Keep rows with non-empty 'text' column
]

filtered_data.to_csv('aapl_sorted_data.csv', index=False)
```

```python
# Step 2: Design the Vocabulary
# Tokenize and create a set of unique words (ignoring case and punctuation)
vocabulary = set()
for text in data['text']:
    words = str(text).lower().split()
    vocabulary.update(words)

vocabulary
```
